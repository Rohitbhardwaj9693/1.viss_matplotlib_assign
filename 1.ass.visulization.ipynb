```python
# Q2: What is a scatter plot? Use the following code to generate data 
# for x and y. Using this generated data plot a scatter plot.
                                       
    
# Scatter plots are the graphs that present the relationship between two
# variables in a data-set.
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
colours = np.random.rand(50)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('this is title,random data')
plt.scatter(x,y,c = colours)
plt.show()
```

```python
# Bar plots are a type of data visualization used to represent data in the
# form of rectangular bars
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.ylabel('company name')
plt.xlabel('profit %')
plt.title('it is a random data')
plt.barh(company,profit)
plt.show()
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
# box1 = np.random.normal(100, 10, 200)
# box2 = np.random.normal(90, 20, 200)
```

```python
# Boxplots are a measure of how well distributed the data in a data set is
# it divides the data set into three quartiles.

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
box = [box1,box2]
fig = plt.figure(figsize =(10, 7))
plt.boxplot(box)
plt.show()
```
